# Analysis

Compare the quality of the lyrics generated by the unigram model (in Lab B) and the bigram model (in Lab C). Which model seems to generate more reasonable lyrics? Can you explain why? What do you see as the advantages and disadvantages of each model?

The bigram model seems to generate more reasonable lyrics compared to the unigram model. This might be because the bigram model stores words for every 2 consecutive words, rather than 1 word per previous word. This makes the random choice more sensible because it chooses words that happen after every two words rather than one word. Choosing a word for every 2 words makes the combination more reasonable. Using the bigram model helps with a more reasonable output as described above, but it could take longer to build the model and it uses more memory to store each tuple of words. The unigram model is less accurate, but it doesn't take as much storage and building it is faster.
